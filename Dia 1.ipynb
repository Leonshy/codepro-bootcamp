## Funciones con parametro y argumentos
las funciones pueden contener parametros y argumentes
los parametros son cuando al llamar a la función y llama a una variable definida se llama parametro
los argumentos son cuanto se llama a la función y en el programa se le envia el valor que va a tener la variable de la función

```python
def preparar_sangüi(x,y):
    print(f"preparar sangüi de {x} con {y}")

ingrediente1=input("escriba tu primer ingrediente")
ingrediente2=input("escriba tu segundo ingrediente")

preparar_sangüi(ingrediente1,ingrediente2)
```
