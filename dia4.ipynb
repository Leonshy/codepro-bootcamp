```python
# Definir una clase madre que herede a sus clases hijas dos atributos de instancia. 
# Mostrar ambos atributos en pantalla desde los objetos de las clases hijas.
# ObservaciÃ³n: Editar las clases hijas para que no se inicialicen.

class Abuela:
    genes = "Matriarcal"
    def __init__(self,nombre,ropa,mueble):
        self.nombre = nombre
        self.ropa = ropa
        self.mueble = mueble
    def consentir(self):
        print(f"{self.nombre} consiente a su nieta regalandole {self.ropa}")    

class Madre(Abuela):
    def retar(self):
        print(f"{self.nombre} reta a su hija por romper el {self.mueble}")
class Hija(Madre):
    def escaparse(self):
        print(f"{self.nombre} agarra el {self.ropa} que le regalo su abuela y se escapa por romper el {self.mueble}")

abuela1 = Abuela("Anatracia", "vestido", "comoda")
madre1 = Madre("Ana", "blusa", "ropero")
hija1 = Hija("Anita", "pantalon para tiendas","mesita de luz")
abuela1.consentir()
madre1.retar()
hija1.escaparse()
```
